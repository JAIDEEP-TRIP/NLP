#PORTER STEMMER

```python

words = ["running", "played", "happiness", "studying", "jumper", 
         "caring", "easily", "playful", "runner", "studied", 
         "jumping", "happier", "teaches", "writing", "carefully"]
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
for word in words:
    print(word + "  >  "+ps.stem(word))
```

```python
from nltk.stem import RegexpStemmer

re = RegexpStemmer('ing$|s$|ed$|er$|ful$|ly$|ness$', min=4)

words = ["running", "played", "happiness", "studying", "jumper", 
         "caring", "easily", "playful", "runner", "studied", 
         "jumping", "happier", "teaches", "writing", "carefully"]

for word in words:
    print(word + "   >  "+ re.stem(word))
```

```python
##Snowball Stemmer
```

```python
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer('english')
```

```python
for word in words:
    print(word +"  > " + snowball.stem(word))
```
