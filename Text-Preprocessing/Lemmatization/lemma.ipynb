```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
```

```python
lemma.lemmatize('going')
```

```python
##by default pos = n - noun
lemma.lemmatize('going',pos='v')
```

```python
words = ["running", "played", "happiness", "studying", "jumper", 
         "caring", "easily", "playful", "runner", "studied", 
         "jumping", "happier", "teaches", "writing", "carefully"]
```

```python
for word in words:
    print(word +"---->"+ lemma.lemmatize(word))
```
