```python
sentence = "Jaideep Tripurani, a software engineer from Denton, Texas, recently visited the Google headquarters in Mountain View, California, and met with the CEO Sundar Pichai on Monday to discuss an AI project worth 5 million dollars."
```

```python
import nltk
from nltk.tokenize import word_tokenize

words = nltk.word_tokenize(sentence)
```

```python
tagged_elements = nltk.pos_tag(words)
```
