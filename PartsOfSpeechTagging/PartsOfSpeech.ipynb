```python
speech = "My fellow citizens, I am standing here in front of you all today because I believe that we as a society have to work together if we are ever going to solve the problems that are affecting each and every one of us. Over the years, I have seen so many of our people struggle with issues that are not only difficult but are also very much avoidable. We have the resources, we have the talent, and we certainly have the will to make things better for all of us. But the question that I keep asking myself is, are we really doing enough? I think that we can do so much more if we just put our minds to it and stop making excuses for the things that we have not yet accomplished. The future of this great nation depends on the choices that you and I make today, not tomorrow, not next week, but right here and right now. So I urge each and every one of you to stand up, speak out, and take action because together there is absolutely nothing that we cannot achieve."
```

```python
import nltk
from nltk.tokenize import sent_tokenize
documents = sent_tokenize(speech)
documents
```

```python
#we will findout the pos tag
from nltk.corpus import stopwords
nltk.download('stopwords')

```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
```

```python
for i in range(len(documents)):
    words = nltk.word_tokenize(documents[i])
    words = [word for word in words if word not in set(stopwords.words('english'))]
    pos_tag = nltk.pos_tag(words)
    print(pos_tag)
```
